# Download/Unload the Nifty, MidcapNifty, BankNifty, FinNifty Options - Calls/Puts for 4 weekly expiry dates using Breeze API

```python
from breeze_connect import BreezeConnect
import numpy as np
import pandas as pd
import time
```

```python
breeze = BreezeConnect(api_key="")
breeze.generate_session(api_secret="", session_token="38879455")
```

```python
# Callback to receive ticks.
def on_ticks(ticks):
    print("Ticks: {}".format(ticks))
   
# Assign the callbacks.
breeze.on_ticks = on_ticks

def truncate(n, decimals = 0):
    multiplier = 10 ** decimals
    return int(n * multiplier) / multiplier
```

```python
def hist_options_data(stock_code,time_interval,start_date,end_date,expiryDates,strikeArray,tempOptionsData):

    for expiry in expiryDates:
        print("for the expiry date",expiry)
        time.sleep(2)
        for n in strikeArray:
            strike = n
            time.sleep(2)
            # downloading historical data for put option contract

            data1 = breeze.get_historical_data(interval = time_interval,
                                    from_date = start_date,
                                    to_date = end_date,
                                    stock_code = stock_code,
                                    exchange_code = "NFO",
                                    product_type = "options",
                                    expiry_date = expiry,
                                    right = "call",
                                    strike_price = strike)

            call_data = pd.DataFrame(data1["Success"])

            data2 = breeze.get_historical_data(interval = time_interval,
                                    from_date = start_date,
                                    to_date = end_date,
                                    stock_code = stock_code,
                                    exchange_code = "NFO",
                                    product_type = "options",
                                    expiry_date = expiry,
                                    right = "put",
                                    strike_price = strike)

            put_data = pd.DataFrame(data2["Success"])
            
            if call_data.shape[0]>0 and put_data.shape[0] > 0:
                optionData = pd.merge(call_data, put_data, on="datetime")
                if(tempOptionsData.shape[0] > 1):
                    tempOptionsData = pd.concat([tempOptionsData,optionData])
                else:
                    tempOptionsData = optionData

                print((strike, tempOptionsData.shape))
    
    return tempOptionsData
```

```python
newColumns = ['datetime', 'stock_code_x', 'exchange_code_x', 'product_type_x',
       'expiry_date_x', 'strike_price_x', 'close', 'right_x', 'open_x', 'high_x',
       'low_x', 'close_x', 'volume_x', 'open_interest_x',
       'right_y', 'open_y', 'high_y', 'low_y', 'close_y',
       'volume_y', 'open_interest_y']

colNames = ['datetime', 'stock_code', 'exchange_code', 'product_type',
   'expiry_date', 'strike_price', 'spot_price','callRight', 'callOpen', 'callHigh',
   'callLow', 'callClose', 'callVolume', 'callOI',
   'putRight', 'putOpen', 'putHigh', 'putLow', 'putClose',
   'putVolume', 'putOI']
```

```python
stock_code = "CNXBAN"
start_date = "2024-01-01T07:00:00.000Z"
end_date = "2024-01-31T18:00:00.000Z"
time_interval = "1minute"
future_data = "NFO_NIFTYBANK_Future_Jan2024" 
filename = "NFO_NIFTYBANK_Options_Jan2024"

startStrike = 47000
endStrike = 50000
print(f"Start and End strike: ({startStrike}, {endStrike})")
strikeArray = range(startStrike, endStrike, 500)

tempOptionsData = pd.DataFrame()

expiryDates = ["2024-01-03","2024-01-10","2024-01-17","2024-01-25"]

monthlyOptionsData = hist_options_data(stock_code,time_interval,start_date,end_date,expiryDates,strikeArray,tempOptionsData)
future_data = pd.read_csv(f"{future_data}.csv")
df_future_options = pd.merge(monthlyOptionsData, future_data, on="datetime")

subOptionsData = df_future_options[newColumns]

subOptionsData.columns = colNames
subOptionsData.to_csv(f"{filename}.csv", index=False)
```
